```python
mplt.figure(figsize=(8,5))
sbrn.histplot(df['attendance_pct'], bins=20)
mplt.title("attendance distribution")
mplt.xlabel("attendance percentage")
mplt.show()
```

```python
mplt.figure(figsize=(8,5))
sbrn.histplot(df['homework_pct'], bins=20)
mplt.title("homework distribution")
mplt.xlabel("home work percentage")
mplt.show()
```

```python
mplt.figure(figsize=(8,5))
sbrn.histplot(df['midterm_score'], bins=20)
mplt.title("midterm scores distribution")
mplt.xlabel("midterm score")
mplt.show()
```

```python
mplt.figure(figsize=(8,5))
sbrn.histplot(df['study_hours_per_week'], bins=20)
mplt.title("study hours / week distribution")
mplt.xlabel("study hours per week")
mplt.show()
```

```python
fig =mplt.figure(figsize=(8,7))
ax = fig.add_subplot(111, projection='3d')
x=df.attendance_pct
y=df.midterm_score
z=df.homework_pct
ax.scatter3D(x, y, z, color='blue', marker='o')
mplt.xlabel("attendance percentage")
mplt.ylabel("midterm score")
ax.set_zlabel("home work percentage")
mplt.title("attendance to medterm score and home work")

mplt.show()
```

```python
fig =mplt.figure(figsize=(8,7))
ax = fig.add_subplot(111, projection='3d')
x=df.study_hours_per_week
y=df.midterm_score
z=df.homework_pct
ax.scatter3D(x, y, z, color='blue', marker='o')
mplt.xlabel("study hours per week")
mplt.ylabel("midterm score")
ax.set_zlabel("home work percentage")
mplt.title("attendance to medterm score and homework percentage")
mplt.show()
```

```python
mplt.figure(figsize=(6,4))
sbrn.heatmap(df[['attendance_pct','homework_pct','midterm_score','study_hours_per_week','pass']].corr(), annot=True)
mplt.title("attendance Correlation Matrix")
mplt.show()
```
