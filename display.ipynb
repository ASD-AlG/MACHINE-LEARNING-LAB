```python
mplt.figure(figsize=(8,5))
sbrn.histplot(df['attendance_pct'], bins=20)
mplt.title("attendance distribution")
mplt.xlabel("attendance percentage")
mplt.show()
```

```python
mplt.figure(figsize=(8,5))
sbrn.histplot(df['homework_pct'], bins=20)
mplt.title("homework distribution")
mplt.xlabel("home work percentage")
mplt.show()
```

```python
mplt.figure(figsize=(8,5))
sbrn.histplot(df['midterm_score'], bins=20)
mplt.title("midterm scores distribution")
mplt.xlabel("midterm score")
mplt.show()
```

```python
mplt.figure(figsize=(8,5))
sbrn.histplot(df['study_hours_per_week'], bins=20)
mplt.title("study hours / week distribution")
mplt.xlabel("study hours per week")
mplt.show()
```

```python
mplt.figure(figsize=(8,5))
sbrn.scatterplot(x='attendance_pct', y='midterm_score', data=df)
mplt.title("attendance to medterm score")
mplt.xlabel("attendance percentage")
mplt.ylabel("midterm score")
mplt.show()
```

```python
mplt.figure(figsize=(8,5))
sbrn.scatterplot(x='attendance_pct', y='homework_pct', data=df)
mplt.title("attendance to home work")
mplt.xlabel("attendance percentage")
mplt.ylabel("homework percentage")
mplt.show()
```

```python
mplt.figure(figsize=(8,5))
sbrn.scatterplot(x='study_hours_per_week', y='midterm_score', data=df)
mplt.title("study hors to medterm score")
mplt.xlabel("study hours per week")
mplt.ylabel("midterm score")
mplt.show()
```

```python
mplt.figure(figsize=(8,5))
sbrn.scatterplot(x='study_hours_per_week', y='homework_pct', data=df)
mplt.title("study hours home work to home work")
mplt.xlabel("study hours per week")
mplt.ylabel("homework percentage")
mplt.show()
```

```python
mplt.figure(figsize=(6,4))
sbrn.heatmap(df[['attendance_pct', 'pass']].corr(), annot=True)
mplt.title("attendance Correlation Matrix")
mplt.show()
```

```python
mplt.figure(figsize=(6,4))
sbrn.heatmap(df[['homework_pct', 'pass']].corr(), annot=True)
mplt.title("home work Correlation Matrix")
mplt.show()
```
