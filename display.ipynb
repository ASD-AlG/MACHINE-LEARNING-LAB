```python
#scaling data frame
df_scaled = df[['Age','daily_screen_time_hours','social_media_hours','online_study_hours','gaming_hours','sleep_hours','attendance_percentage','offline_study_hours','previous_sem_CGPA','current_sem_CGPA']].copy()
df_scaled[['Age','daily_screen_time_hours','social_media_hours','online_study_hours','gaming_hours','sleep_hours','attendance_percentage','offline_study_hours','previous_sem_CGPA','current_sem_CGPA']] = mmscaler.fit_transform(df_scaled)
df_scaled.head()
```

```python
#normalizing data frame
df_standardized = df_scaled[['Age','daily_screen_time_hours','social_media_hours','online_study_hours','gaming_hours','sleep_hours','attendance_percentage','offline_study_hours','previous_sem_CGPA','current_sem_CGPA']].copy()
df_standardized[['Age','daily_screen_time_hours','social_media_hours','online_study_hours','gaming_hours','sleep_hours','attendance_percentage','offline_study_hours','previous_sem_CGPA','current_sem_CGPA']] = scaler.fit_transform(df_standardized)
df_standardized.head()
```

```python
#new corralation map
mplt.figure(figsize=(10,5))
sbrn.heatmap(df_standardized[['Age','daily_screen_time_hours','social_media_hours','online_study_hours','gaming_hours','sleep_hours','attendance_percentage','offline_study_hours','previous_sem_CGPA','current_sem_CGPA']].corr(), annot=True)
mplt.title("Correlation Matrix")
mplt.show()
#same map so values didnt get affected
```

```python
X = df_standardized[['Age','daily_screen_time_hours','social_media_hours','online_study_hours','gaming_hours','sleep_hours','attendance_percentage','offline_study_hours','previous_sem_CGPA','current_sem_CGPA']]

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)

print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```
